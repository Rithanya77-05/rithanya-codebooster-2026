```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("All libraries are important")
```

```python
df=pd.read_csv("/content/student_performance.csv")
print(f"Data loaded: {df.shape[0]} students, {df.shape[1]} columns")
print(f"Columns: {df.columns.tolist()}")
print(f"Missing values: {df.isnull().sum().sum()}")
df.head()
```

```python
df_ml = df.copy()
le_gender=LabelEncoder()
df_ml["gender_coded"]=le_gender.fit_transform(df_ml["gender"])
print(f"Gender encoding:{dict(zip(le_gender.classes_,le_gender.transform(le_gender.classes_)))}")
le_dept=LabelEncoder()
df_ml["Department_encoded"]=le_dept.fit_transform(df_ml["department"])
print(f"Department encoding: {dict(zip(le_dept.classes_,le_dept.transform(le_dept.classes_)))}")
print()
```

```python
feature_cols=[
    'math_score',
    'science_score',
    'english_score',
    'gender_coded',
    'Department_encoded'
]
x=df_ml[feature_cols]
y=df_ml['programming_score']

print(f"Feature matrix shape: {x.shape}")
print(f"Target vector shape: {y.shape}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f"Total students: {len(x)}")
print(f"Training students: {len(x_train)} ({len(x_train)/len(x)*100:.2f}%)")
print(f"Testing students: {len(x_test)} ({len(x_test)/len(x)*100:.2f}%)")
print(f"\nTraining target range: {y_train.min()} to {y_train.max()}")
print(f"Testing target range: {y_test.min()} to {y_test.max()}")
#
```

```python
lr_model = LinearRegression()

# Combine x and y for consistent NaN dropping
train_combined = pd.concat([pd.DataFrame(x_train), y_train], axis=1).dropna()
x_train_cleaned = train_combined[feature_cols]
y_train_cleaned = train_combined['programming_score']

test_combined = pd.concat([pd.DataFrame(x_test), y_test], axis=1).dropna()
x_test_cleaned = test_combined[feature_cols]
y_test_cleaned = test_combined['programming_score']

# Scale the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_cleaned)
x_test_scaled = scaler.transform(x_test_cleaned)

# Fit the model using cleaned and scaled data
lr_model.fit(x_train_scaled, y_train_cleaned)

# Make predictions
lr_pred = lr_model.predict(x_test_scaled)

# Calculate evaluation metrics (corrected indentation)
lr_mae = mean_absolute_error(y_test_cleaned, lr_pred)
lr_mse = mean_squared_error(y_test_cleaned, lr_pred)
lr_r2 = r2_score(y_test_cleaned, lr_pred)

# Print results
print(f"Linear Regression - MAE: {lr_mae:.2f}, MSE: {lr_mse:.2f}, R2: {lr_r2:.2f}")
print(f"RMSE: {np.sqrt(lr_mse):.2f}")
print(f"Coefficients: {lr_model.coef_}")
print(f"Intercept: {lr_model.intercept_}")
```

```python
dt_model = DecisionTreeRegressor(
    max_depth=5,
    random_state=42
)
dt_model.fit(x_train_cleaned,y_train_cleaned)
dt_pred=dt_model.predict(x_test_scaled)
dt_mae=mean_absolute_error(y_test_cleaned,dt_pred)
dt_rmse=np.sqrt(mean_squared_error(y_test_cleaned,dt_pred))
dt_r2=r2_score(y_test_cleaned,dt_pred)
print(f'Decision Tree MAE:{dt_mae:.2f}')
print(f'Decision Tree RMSE:{dt_rmse:.2f}')
print(f'Decision Tree R2:{dt_r2:.2f}')
```
