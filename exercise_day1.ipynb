```python
import pandas as pd
```

```python
df = pd.read_csv('/content/student_performance.csv')
```

```python
print("DATASET OVERVIEW")
print("-"*40)
print("Total number of students:",df.shape[0])
print("Total number of columns:",df.shape[1])
print("Departments:",len(df['department'].unique()))
```

```python
print(df['department'].value_counts())
```

```python
print("Highest math score:", df['math_score'].max())
print("Lowest score:",df['math_score'].min())
print("AVerage math score:",df['math_score'].mean().round(2))
print("AVerage Programming score:",df['programming_score'].mean().round(2))
print("AVerage attendance:",df['attendance_percentage'].mean().round(2))
```

```python
df['total_score']=df['math_score']+df['science_score']+df['english_score']+df['programming_score']
```

```python
gender_scores=df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)
```

```python
top5=df.sort_values('total_score', ascending=False).head(5)
for index, row in top5.iterrows():
  print(f" {row['name']:<20} | {row['department']:<18} | Score:{row['total_score']}")
```

```python
low_att=df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_att)>0:
  print(low_att.to_string(index=False))
else:
  print("All Students have the attendance above 75%")
```
