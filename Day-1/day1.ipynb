```python
cs_students=df[df['department']=='Computer Science']
print(f"number of Coomputer Science students: {len(cs_students)}")
print("\nComputer Science students:")
print(cs_students[['name','math_score', 'programming_score','attendance_percentage']])
```

```python
high_attendance_students=cs_students[cs_students['attendance_percentage'] > 90]
print(high_attendance_students[['name','attendance_percentage']])
```

```python
df['total_score']=df['math_score']+df['science_score']+df['english_score']+df['programming_score']
toppers=df[df['total_score']/4 >90]
print(toppers[['name','total_score']])
```

```python
dept_avg=df.groupby('department')['total_score'].mean().round(2)
print("Average total score by department:")
print(dept_avg)
```

```python
print("Average total score by department (Descending):")
print(dept_avg.sort_values(ascending=False))
```
