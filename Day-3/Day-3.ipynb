```python
raw_df=pd.read_csv('/content/messy_sales_data.csv')
print(f"Raw data loaded:  {raw_df.shape[0]} rows,{raw_df.shape[1]} columns")
print(f"columns in the dataset :{raw_df.columns.tolist()}")
print("First 5 rows")
print(raw_df.head().to_string(index=False))
```

```python
print('\n MISSING VALUES per column:')
print(raw_df.isnull().sum())
print(f"\n DuPLICATE ROWS: {raw_df.duplicated().sum()}")
print('\n Data types of each column:')
print(raw_df.dtypes)
print(f"Unique categories: {raw_df['category'].unique()} ")
print(f" Sample customer nmaes:", raw_df['customer_name'].dropna().unique()[:8]) #dropna it will remove the null value and disply from the next
print(" sample oder_date", raw_df['order_date'].unique()[:6])
```

```python
df=raw_df.copy() #gonna work in this copy datset only
print(f"Working copy created:{df.shape}")
print('Raw_df is untouched - we can always reet by running df= raw_df.copy()')
```

```python
print('Before fixing nulls:', df.isnull().sum().sum(),'total missing values') #total missing values it will give for second .sum()
df['customer_name'].fillna('Unknown Customer', inplace=True) # fillna() meaning it will replace the null customer with unkown customer
median_qty=df['quantity'].median() #take the median value
df['quantity'].fillna(median_qty, inplace=True)
print(f"Filled misssing quantity with median: {median_qty}")
df['category'].fillna("Uncategorized",inplace=True)
print("After fixing values:", df.isnull().sum().sum(), 'Total missing values')
print(df.head().to_string(index=False))
df['product'].isnull().sum()
df['product'].fillna('Mysteeerious product',inplace=True)
df.head()
#df['product'].fillna()
```

```python
## REmoving duplicates
print(f'Be4 duplicate: {len(df)} rows')
print(f'duplicate rows: {df.duplicated().sum()}')
print("\n duplicate rows:")
print(df[df.duplicated(keep=False)][['order_id', 'customer_name','product']]) ## keep=False
df.drop_duplicates(inplace=True) ## in raw_df alsoo we are going to store it,, not only in df

print(f'\nAfter duplication: {len(df)} rows')
print(f'Rows removed: {len(raw_df)-len(df)}')
```

```python
print('sample dates before parsing')
print(df['order_date'].head(8))
df['order_date']= pd.to_datetime(
    df['order_date'],
    dayfirst=False,  #no need of day first
    errors='coerce' # it will fix itself and then display
)

nat_count = df['order_date'].isnull().sum()
print(f"\n Unparaseable dates(NaT): {nat_count}") #not a time nat

df['year'] = df['order_date'].dt.year
df['month']=df['order_date'].dt.month
df['month_name']= df['order_date'].dt.strftime('%B')
df['day_name']=df['order_date'].dt.strftime('%A')

print('\n Sample dates after paring:')
print(df[['order_date','year','month','month_name','day_name']].head(5))
```

```python
print(f"Before standardization: {df['customer_name'].tolist()[:5]}" ) #tolist () and unique() are also displaying in list format
df['customer_name']=(
    df['customer_name']
    .str.title()
    .str.strip()
)
print(f"After standartization: {df['customer_name'].tolist()[:5]}")
```

```python
print('Revenue Column created:')
df['revenue']= df['quantity']*df['unit_price']
print(df[ ['customer_name','quantity','unit_price','revenue']].head(10))
print("Total revenue",df['revenue'].sum())
```

```python
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce').astype(int)
df.head(30)
```
