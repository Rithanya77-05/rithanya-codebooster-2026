````python
import pandas as pd
````

````python
df = pd.read_csv('/content/student_performance.csv')
````

```markdown
## 3. Department-wise Analysis

Let's see how students are distributed across various departments.
```

````python
# Display the count of students in each department
department_counts = df['department'].value_counts()

print("Number of students in each department:")
print(department_counts)
````

```markdown
## 4. Score Analysis

Let's analyze the various score metrics for the students.
```

````python
# Find and Display:
# - highest math score
# - lowest math score
# - avg math score
# - avg programming score
# - avg attendance percentage

highest_math_score = df['math_score'].max()
lowest_math_score = df['math_score'].min()
avg_math_score = df['math_score'].mean()
avg_programming_score = df['programming_score'].mean()
avg_attendance_percentage = df['attendance_percentage'].mean()

print(f"Highest Math Score: {highest_math_score}")
print(f"Lowest Math Score: {lowest_math_score}")
print(f"Average Math Score: {avg_math_score:.2f}")
print(f"Average Programming Score: {avg_programming_score:.2f}")
print(f"Average Attendance Percentage: {avg_attendance_percentage:.2f}%")
````

```markdown
## 5. Gender-wise Analysis

Let's calculate and display the average total score by gender.
```

````python
# Calculate total_score if it doesn't exist, by summing math_score and programming_score
if 'total_score' not in df.columns:
    df['total_score'] = df['math_score'] + df['programming_score']

# Calculate and display average total score by gender
avg_total_score_by_gender = df.groupby('gender')['total_score'].mean()

print("Average Total Score by Gender:")
print(avg_total_score_by_gender)
````

```markdown
## 6. Top Performers

Let's identify the top 5 students based on their total score.
```

````python
# Display top 5 students based on total score.
# Show:
# - Student name
# - Department
# - Total score

top_5_performers = df.sort_values(by='total_score', ascending=False).head(5)

print("Top 5 Students by Total Score:")
display(top_5_performers[['name', 'department', 'total_score']])
````

```markdown
## 7. Attendance Analysis

Let's find students whose attendance is below 75%.
```

````python
# Find students whose attendance is below 75%
students_low_attendance = df[df['attendance_percentage'] < 75]

if not students_low_attendance.empty:
    print("Students with attendance below 75%:")
    display(students_low_attendance[['name', 'department', 'attendance_percentage']])
else:
    print("All students have attendance above 75%.")
````
